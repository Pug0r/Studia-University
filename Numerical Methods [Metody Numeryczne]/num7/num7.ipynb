```python
import numpy as np
from scipy.interpolate import interp1d
from scipy.stats import norm
import matplotlib.pyplot as plt
from math import cos, pi

def func1(x: float):
    return 1 / (1 + 25*x*x)

def func2(x: float):
    return x*x*x

def func3(x: float):
    return norm.pdf(x)

def equal_split(n: int,  i: int):
    return -1 + 2*i/(n+1)

def cos_split(n: int, i: int):
     return cos(((2*i+1)/(2*(n+1)) )*pi)
    
def interpolate(x: float, coords: list, values: list)-> list:
    n = len(coords)
    result = 0
    for j in range(n):
        num = 1
        den = 1
        for i in range(n):
            if (j != i):
                num *= x - coords[i]
                den *= coords[j] - coords[i]
        result += (num/den) * values[j]
    return result


def gen_cords_values(n: int, split: object, func: object):
    coords = []
    values = []
    for i in range(n):
        temp = split(n, i)
        coords.append(temp)
        values.append(func(temp))
    return coords, values
```

```python
# Equal split
# Let's set n = 10, n=20, n=50, n=100
draw_points = np.linspace(-1, 1)

fig, ax = plt.subplots(figsize=(20, 10))
ax.set_title('Function 1/(1 + 25*x^2) interpolation at the equal split')
ax.set_ylim([-1.5, 1.5])

n = 10
x, y = gen_cords_values(n, equal_split, func1)
draw_values = []
for argument in draw_points:
    draw_values.append(interpolate(argument, x, y))
    
ax.plot(draw_points, draw_values, label='n=10');
ax.scatter(x, y);

n = 20
x, y = gen_cords_values(n, equal_split, func1)
draw_values = []
for argument in draw_points:
    draw_values.append(interpolate(argument, x, y))
    
ax.plot(draw_points, draw_values, label='n=20');

n = 50
x, y = gen_cords_values(n, equal_split, func1)
draw_values = []
for argument in draw_points:
    draw_values.append(interpolate(argument, x, y))
    
ax.plot(draw_points, draw_values, label='n=50');

n = 100
x, y = gen_cords_values(n, equal_split, func1)
draw_values = []
for argument in draw_points:
    draw_values.append(interpolate(argument, x, y))
    
ax.plot(draw_points, draw_values, label='n=100');


ax.legend();
plt.savefig('fun1_eq.pdf')
```

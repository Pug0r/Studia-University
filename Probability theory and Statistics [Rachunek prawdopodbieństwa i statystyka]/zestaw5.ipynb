```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.stats import chi2

def gen_nor_dist(u1: list, u2: list):
    z1 = []
    z2 = []
    for x, y in zip(u1, u2):
        z1.append(np.sqrt(-2*np.log(x)) * np.cos(2*np.pi*y))
        z2.append(np.sqrt(-2*np.log(y)) * np.cos(2*np.pi*x))
    return z1, z2
```

```python
var1 = np.random.uniform(size=10**6)
var2 = np.random.uniform(size=10**6)
var3 = np.random.uniform(size=10**6)
var4 = np.random.uniform(size=10**6)

norm1, norm2 = gen_nor_dist(var1, var2)
norm3 = gen_nor_dist(var4, var3)[0]
chi_square = [x**2 + y**2 + z**2 for x,y,z in zip(norm1, norm2, norm3)]
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

x = np.linspace(0, 20, 10000)

ax.plot(x, chi2.pdf(x, 3));
plt.hist(chi_square, bins=55, density=True);
plt.savefig('zestaw5_poprawione.pdf')
```

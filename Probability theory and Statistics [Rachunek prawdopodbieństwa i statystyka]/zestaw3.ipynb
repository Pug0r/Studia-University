```python
import random 
from math import sqrt
import matplotlib.pyplot as plt 
```

```python
def rev_dist_fun(argument):
    if argument <= 0 or argument > 1:
        raise ValueError('Argument must be in range (0;1>')
    elif argument > 0 and argument <= (1/6):
        return sqrt(6*argument)-1
    elif argument > (1/6) and argument <= (5/6):
        return 3*argument-(1/2)
    elif argument > (5/6):
        return 3-sqrt(6-6*argument)
    
def dens_func(argument):
    if argument < -1:
        return 0
    elif argument > -1 and argument < 0:
        return (1/3)*argument+(1/3)
    elif argument >= 0 and argument < 2:
        return 1/3
    elif argument >= 2 and argument < 3:
        return (-1/3)*argument+1
    else:
        return 0
    
```

```python
gen_num = (10**2, 10**3, 10**4, 10**5, 10**6)
results = []

for number in gen_num:
    rnd = [random.random() for _ in range(number)]
    x_arg = [rev_dist_fun(x) for x in rnd]
    y_arg = [dens_func(x) for x in x_arg]
    results.append((x_arg, y_arg))
    
```

```python
import matplotlib.patches as patches

x_cornes = (-1, 3, 2, 0)
y_corners = (0, 0, 1/3, 1/3)

for i in range(4):
    fig, ax = plt.subplots(figsize=(20, 10))
    ax.bar(results[i][0], results[i][1], width=0.05);
    ax.add_patch(patches.Polygon(xy=list(zip(x_cornes,y_corners)), fill=False, color='red'));
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.set_title(f'Density function for {gen_num[i]} points')
    plt.savefig(f'plot{i}.pdf')
```

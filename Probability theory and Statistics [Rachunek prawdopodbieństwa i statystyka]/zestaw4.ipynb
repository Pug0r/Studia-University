Rozwiązanie analityczne funkcji gęstości prowadzi do ogólnego wzoru F(y) = 1 / (2*n*y**(1 - 1/n)) dla n wymiarów. 
Wykresy zgadzają się z intuicją, że będziemy obserwować tym mniej (rzadziej rozstawionych) słupków im mniej prawdopodobne będzie otrzymana objętość (skrajnie wielka lub skrajnie mała).

Można zaobserwować, że objętości są raczej ściągane w dół niż w gorę, z tego względu, że coraz mniejszy przedział z zakresu (0,2) odpowiada za (potencjalne) osiągnięcie coraz większych objętości - więc jest to coraz mniej prawdopodobnę

Dla n=20 zmieniłem ilość generowanych punktów, przedział oraz skalę - inaczej wykres był niezbyt czytelny (a i tak szału nie ma)

```python
import matplotlib.pyplot as plt
import numpy as np
from random import random


def density_function(n: int, x: float):
    return 1 / (2*n*x**(1 - 1/n))

def numeric_function(dim: int):
    volume = 1
    for _ in range(dim):
        volume *= random()*2
    return volume
```

```python
N = 1
POINTS_GEN = 1000

points = np.linspace(0, 2**N, num=POINTS_GEN)
points_values = [density_function(N, x) for x in points]
num_points = [numeric_function(N) for _ in range(POINTS_GEN)]
num_points_values = [density_function(N, x) for x in num_points]

fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(points, points_values, color='red');
ax.bar(num_points, num_points_values, width=0.01);
```
